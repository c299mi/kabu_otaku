```python
#日本証券所のウェブサイトから全上場銘柄のシートを取得
#https://www.jpx.co.jp/markets/statistics-equities/misc/01.html

import pandas as pd

# CSVファイルのパス
file_path = 'data/data_j.csv'  # 実際のファイルパスに置き換えてください

# CSVファイルの読み込み
df = pd.read_csv(file_path)

def extract_market_tickers(df, market_name, output_file_path):
    """
    指定した市場区分の銘柄コードを抽出し、指定されたCSVファイルに保存する関数
    """
    # 指定した市場区分の行を抽出
    market_df = df[df['市場・商品区分'] == market_name]

    # 銘柄コードを抽出
    market_tickers = market_df['コード']

    # 抽出した銘柄コードを表示
    print(f"{market_name}の銘柄コード:")
    print(market_tickers.tolist())

    # CSVファイルに保存
    market_tickers.to_csv(output_file_path, index=False, header=True)
    print(f"{market_name}の銘柄コードを {output_file_path} に保存しました。")

# グロース市場の銘柄コードを抽出し、指定したファイル名で保存
extract_market_tickers(df, 'グロース（内国株式）', 'data/growth_tickers.csv')

# スタンダード市場の銘柄コードを抽出し、指定したファイル名で保存
extract_market_tickers(df, 'スタンダード（内国株式）', 'data/standard_tickers.csv')

# プライム市場の銘柄コードを抽出し、指定したファイル名で保存
extract_market_tickers(df, 'プライム（内国株式）', 'data/prime_tickers.csv')


```
